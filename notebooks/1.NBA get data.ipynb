### Importing necessary packages

```python
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.endpoints import boxscoreadvancedv2
from nba_api.stats.endpoints import boxscoreusagev2
import pandas as pd
```

### Getting initial seasons basic box scores

```python
seasons = ["2018-19","2019-20", "2020-21", "2021-22", "2022-23"]
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable='2017-18',
                                               league_id_nullable='00',
                                               season_type_nullable='Regular Season')
games = gamefinder.get_data_frames()[0]

print("2017-18", len(games))


for season in seasons:
    gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable=season,
                                               league_id_nullable='00',
                                               season_type_nullable='Regular Season')
    temp = gamefinder.get_data_frames()[0]
    print(season, len(temp))
    
    games = pd.concat([games, temp]).reset_index(drop=True)
```

```python
games
```

```python
games.columns
```

#### Getting ids from games

```python
ids = []
seen = {}

for i in range(len(games)):
    if games['GAME_ID'][i] not in seen:
        seen[games['GAME_ID'][i]] = 1
        ids.append(games['GAME_ID'][i])
```

### Getting advanced box scores for initial seasons

### Repeating process for playoff games in initial seasons

```python
gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable='2017-18',
                                               league_id_nullable='00',
                                               season_type_nullable='Playoffs')
gamesP = gamefinder.get_data_frames()[0]

print("2017-18", len(gamesP))


for season in seasons:
    gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable=season,
                                               league_id_nullable='00',
                                               season_type_nullable='Playoffs')
    temp = gamefinder.get_data_frames()[0]
    print(season, len(temp))
    
    gamesP = pd.concat([gamesP, temp]).reset_index(drop=True)
```

```python
gamesP
```

```python
idsP = []
seen = {}

for i in range(len(gamesP)):
    if gamesP['GAME_ID'][i] not in seen:
        seen[gamesP['GAME_ID'][i]] = 1
        idsP.append(gamesP['GAME_ID'][i])
```

```python
advP = boxscoreadvancedv2.BoxScoreAdvancedV2(end_period=4,
                                     end_range=0,
                                     game_id=idsP[0],
                                     range_type=0,
                                     start_period=1,
                                     start_range=0).get_data_frames()[1]

for i in range(1, len(idsP)):
    
    temp = boxscoreadvancedv2.BoxScoreAdvancedV2(end_period=4,
                                     end_range=0,
                                     game_id=idsP[i],
                                     range_type=0,
                                     start_period=1,
                                     start_range=0).get_data_frames()[1]
    
    advP = pd.concat([advP, temp]).reset_index(drop=True)
    
    if i%25 == 0:
        print(i)
```

```python
advP
```

```python
playoffs = pd.merge(gamesP, advP, how="inner", left_on=['GAME_ID','TEAM_ID'], right_on = ['GAME_ID','TEAM_ID'])
```

```python
playoffs
```

```python
playoffs.to_csv('playoffs.csv')
```
