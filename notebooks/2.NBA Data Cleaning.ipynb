### Getting DNP players, and making column for game number

```python
## Counting how many key players did not play for a team in each game

dnp = usage[usage['COMMENT'].notna()].reset_index(drop=True)

key_ids = usg_allstar['PLAYER_ID'].unique()

key_dnp = {}

for i in range(len(dnp)):
    if dnp['PLAYER_ID'][i] in key_ids:
        if (dnp['GAME_ID'][i], dnp['TEAM_ID'][i]) in key_dnp:
            key_dnp[(dnp['GAME_ID'][i], dnp['TEAM_ID'][i])] += 1
        else:
            key_dnp[(dnp['GAME_ID'][i], dnp['TEAM_ID'][i])] = 1
    

```

```python
## Assigning values based on team and game id for key DNP players

alldata["KEY_DNP"] = 0

for i in range(len(alldata)):
    if (alldata["GAME_ID"][i], alldata['TEAM_ID'][i]) in key_dnp:
        alldata['KEY_DNP'][i] = key_dnp[(alldata["GAME_ID"][i], alldata['TEAM_ID'][i])]
```

```python
## Making dictionary to keep track of game count
abr = list(alldata.TEAM_ABBREVIATION.unique())
team_game_count = {}

for ab in abr:
    team_game_count[ab] = 1

season = alldata['SEASON_ID'][0]
```

```python
## Creating game count
alldata["GAME_NUMBER"] = 0

for i in range(len(alldata)):
    if alldata['SEASON_ID'][i] != season and alldata['PLAYOFF_IND'][i] !=1:
        season = alldata['SEASON_ID'][i]
        for ab in abr:
            team_game_count[ab] = 1
    alldata['GAME_NUMBER'][i] = team_game_count[alldata['TEAM_ABBREVIATION'][i]]
    team_game_count[alldata['TEAM_ABBREVIATION'][i]] += 1
```

### Creating home team indicator column, and changing W/L to 1s and 0s, and creating new column for win percentage

```python
home = []

for i in range(len(alldata_total)):
    if alldata_total['MATCHUP'][i][4] == "v":
        home.append(1)
    else:
        home.append(0)
        
alldata_total["HOME"] = home

win = []

for i in range(len(alldata_total)):
    if alldata_total['WL'][i] == "W":
        win.append(1)
    else:
        win.append(0)

alldata_total['WL'] = win

alldata_total["WIN_PERC"] = alldata_total['WL']
```

### Creating columns for rest time, and indicator for playing a back to back

```python
## Back to back

back2back = []
rest = []
abr = list(alldata_total.TEAM_ABBREVIATION.unique())
dates = {}

for team in abr:
    dates[team] = datetime.datetime.strptime('2016-01-01', "%Y-%m-%d")
    
for i in range(len(alldata_total)):
    
    if alldata_total['GAME_NUMBER'][i] == 1:
        rest.append(5)
        dates[alldata_total['TEAM_ABBREVIATION'][i]] = datetime.datetime.strptime(alldata_total['GAME_DATE'][i], "%Y-%m-%d")
        back2back.append(0)
    else:
        diff = (datetime.datetime.strptime(alldata_total['GAME_DATE'][i], "%Y-%m-%d") - dates[alldata_total['TEAM_ABBREVIATION'][i]]).days
        if diff == 0:
            print(datetime.datetime.strptime(alldata_total['GAME_DATE'][i], "%Y-%m-%d"))
            print(dates[alldata_total['TEAM_ABBREVIATION'][i]])
            print(i)
            break 
        if diff == 1:
            back2back.append(1)
        else:
            back2back.append(0)
        rest.append(diff)
        dates[alldata_total['TEAM_ABBREVIATION'][i]] = datetime.datetime.strptime(alldata_total['GAME_DATE'][i], "%Y-%m-%d")
        
```

```python
alldata_total['BACK_BACK'] = back2back
alldata_total['REST'] = rest
```
